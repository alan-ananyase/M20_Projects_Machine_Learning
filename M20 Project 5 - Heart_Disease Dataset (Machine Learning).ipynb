```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset
cols=['Age','Sex','Chest_Pain_Type','trest_bps','chol','fasting_blood_sugar','rest_ecg','max_heart_rate','exang','ST_depression','slope','ca','thal','num']
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data', header=None, names=cols)
df
```

There are only 6 values in total that has '?' in 'ca' and 'thal'. We can drop them and convert the columns to int.

```python
a=df[df['thal']!='?']
df=a[a['ca']!='?']
df
```

```python
# Let's covert string values to integers. Since we have '0.0', we need to first convert them to float and then to integer
for i in df.columns:
    if df[i].dtype=='object':
        df[i]=df[i].astype(float).astype(int)
```

```python
# We just need to identify if the patient has heart disease or not. Any non-zero value means, they have heart disease.
df['num']=df.num.map({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})
```

```python
# Splitting into trainng and test set
x=df.drop(columns=['num'])
y=df['num']
print('x.shape:',x.shape,'y.shape:',y.shape)
```

Our dataset is now ready for the next stage > Scaling

```python
#Scaling data using StandardScaler
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20,stratify=y)
print('x_train.shape: ',x_train.shape,'x_test.shape:',x_test.shape,'\ny_train.shape:',y_train.shape,'y_test.shape:',y_test.shape)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,confusion_matrix
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

logr=LogisticRegression(random_state=42)
```

```python
#Gaussian NB
from sklearn.naive_bayes import GaussianNB

gnb=GaussianNB()
```

```python
#Support Vector Classifier
from sklearn.svm import SVC

svc=SVC()
svc_parameters={'kernel':['linear','sigmoid','poly','rbf'],'C':[1,10]}
bsvc=GridSearchCV(svc,svc_parameters)
bsvc.fit(x_train,y_train)
bsvc.best_params_
```

```python
svc=SVC(kernel='linear',C=10)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()
neighbors={'n_neighbors':range(1,30)}
bknn=GridSearchCV(knn,neighbors)
bknn.fit(x_train,y_train)
bknn.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=14)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

criterion = {'criterion':['gini','entropy']}
dtc=DecisionTreeClassifier(random_state=42)
bdtc=GridSearchCV(dtc,criterion)
bdtc.fit(x_train,y_train)
bdtc.best_params_
```

```python
dtc=DecisionTreeClassifier(criterion='gini',random_state=42)
```

```python
#Random Forest Classifer
from sklearn.ensemble import RandomForestClassifier

parameters = {'n_estimators':range(50,251,50)}
rf=RandomForestClassifier(random_state=42)
brf=GridSearchCV(rf,parameters)
brf.fit(x_train,y_train)
brf.best_params_
```

```python
rf=RandomForestClassifier(n_estimators=250,random_state=42)
```

```python
#Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

gbc=GradientBoostingClassifier(n_estimators=250)
```

```python
#Extra Trees Classifier
from sklearn.ensemble import ExtraTreesClassifier

etc=ExtraTreesClassifier(n_estimators=250)
```

```python
#AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier

ada=AdaBoostClassifier(n_estimators=50)
```

```python
#Bagging Classifier
from sklearn.ensemble import BaggingClassifier

bc=BaggingClassifier(n_estimators=250)
```

```python
# XGBoost Classifier
from xgboost import XGBClassifier

xgb=XGBClassifier(n_estimators=250)
```

```python
models={
    'LogisticRegression': logr,
    'Gaussian NB': gnb,
    'SupportVectorClassifer': svc,
    'KNeighborsClassifier': knn,
    'DecisionTreeClassifier': dtc,
    'RandomForestClassifier': rf,
    'GradientBoostingClassifier': gbc,
    'ExtraTreesClassifier': etc,
    'AdaBoostClassifier': ada,
    'BaggingClassifier': bc,
    'XGBClassifier': xgb
}
```

```python
models_name,ascore,roc,cross_val=[],[],[],[]
for name in models.keys():
    models[name].fit(x_train,y_train)
    models_name.append(name)
    y_pred=models[name].predict(x_test)
    ascore.append(round(accuracy_score(y_test,y_pred)*100,2))
    z=(cross_val_score(models[name],x,y,cv=10,scoring='accuracy').mean())
    cross_val.append(round(z*100,2))
    roc.append(round(roc_auc_score(y_test,y_pred)*100,2))
    print('################# ',name,' ## COMPLETED ##')
final=pd.DataFrame({'Model':models_name,'Accuracy Score':ascore,'ROC_AUC_Score':roc,'Cross Val Score':cross_val}).sort_values(by='Accuracy Score',ascending=False)
final
```

Clearly, Gaussian NB gives the best accuracy score of 88.33%

```python
gnb.fit(x_train,y_train)
y_pred=gnb.predict(x_test)
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import roc_curve

y_pred_prob=gnb.predict_proba(x_test)[:,1]  #1 is the probabilty of threshold value
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)

plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Gaussian NB')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression')
plt.show()
```
