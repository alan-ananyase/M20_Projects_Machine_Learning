We learned a lot of details from the above visualization. Let's process the data for model learning.

```python
for name in x.columns:
    if x[name].nunique() <=2:
        if x[name].dtype=='object':
            print(name,'---->',x[name].nunique())
```

We can covert all categorical data with 2 or less categories to numberical data using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for name in x.columns:
    if x[name].nunique() <=2:
        if x[name].dtype=='object':
            x[name]=le.fit_transform(x[name])
            print(name,'----> ## COMPLETED ##')
            print(x[name].value_counts())
```

```python
# Converting remaining categorical variables in x using get_dummies
x=pd.get_dummies(x)
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
y=le.fit_transform(y)
y.shape
```

Let's split training and testing set. Since we have more values in one section of target, we can use stratify to maintain the same ratio.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20,stratify=y)
print('x_train.shape: ',x_train.shape,'x_test.shape:',x_test.shape,'\ny_train.shape:',y_train.shape,'y_test.shape:',y_test.shape)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,confusion_matrix
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
#SVC
from sklearn.svm import SVC

svc=SVC()
svc_parameters={'kernel':['linear','sigmoid','poly','rbf'],'C':[1,10]}
bsvc=GridSearchCV(svc,svc_parameters)
bsvc.fit(x_train,y_train)
bsvc.best_params_
```

```python
svc=SVC(kernel='linear',C=1)
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
svc_score=round(accuracy_score(y_test,y_pred)*100,2)
print('svc_score:',svc_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()
neighbors={'n_neighbors':range(1,30)}
bknn=GridSearchCV(knn,neighbors)
bknn.fit(x_train,y_train)
bknn.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
knn_score=round(accuracy_score(y_test,y_pred)*100,2)
print('knn_score:',knn_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

criterion = {'criterion':['gini','entropy']}
dtc=DecisionTreeClassifier(random_state=42)
bdtc=GridSearchCV(dtc,criterion)
bdtc.fit(x_train,y_train)
bdtc.best_params_
```

```python
dtc=DecisionTreeClassifier(criterion='gini',random_state=42)
dtc.fit(x_train,y_train)
y_pred=dtc.predict(x_test)
dtc_score=round(accuracy_score(y_test,y_pred)*100,2)
print('dtc_score:',dtc_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#Random Forest Classifer
from sklearn.ensemble import RandomForestClassifier

parameters = {'n_estimators':range(50,251,50)}
rf=RandomForestClassifier(random_state=42)
brf=GridSearchCV(rf,parameters)
brf.fit(x_train,y_train)
brf.best_params_
```

```python
rf=RandomForestClassifier(n_estimators=150,random_state=42)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
rf_score=round(accuracy_score(y_test,y_pred)*100,2)
print('rf_score:',rf_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

gbc=GradientBoostingClassifier(n_estimators=250)
gbc.fit(x_train,y_train)
y_pred=gbc.predict(x_test)
gbc_score=round(accuracy_score(y_test,y_pred)*100,2)
print('gbc_score:',gbc_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#Extra Trees Classifier
from sklearn.ensemble import ExtraTreesClassifier

etc=ExtraTreesClassifier(n_estimators=250)
etc.fit(x_train,y_train)
y_pred=etc.predict(x_test)
etc_score=round(accuracy_score(y_test,y_pred)*100,2)
print('etc_score:',etc_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier

ada=AdaBoostClassifier(n_estimators=50)
ada.fit(x_train,y_train)
y_pred=ada.predict(x_test)
ada_score=round(accuracy_score(y_test,y_pred)*100,2)
print('ada_score:',ada_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
#Bagging Classifier
from sklearn.ensemble import BaggingClassifier

bc=BaggingClassifier(n_estimators=250)
bc.fit(x_train,y_train)
y_pred=bc.predict(x_test)
bc_score=round(accuracy_score(y_test,y_pred)*100,2)
print('bc_score:',bc_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
# XGBoost Classifier
from xgboost import XGBClassifier

xgb=XGBClassifier(n_estimators=250)
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
xgb_score=round(accuracy_score(y_test,y_pred)*100,2)
print('xgb_score:',xgb_score,'%\nroc_auc_score:',roc_auc_score(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True)
```

```python
models={
    'SVC': svc,
    'KNeighborsClassifier': knn,
    'DecisionTreeClassifier': dtc,
    'RandomForestClassifier': rf,
    'GradientBoostingClassifier': gbc,
    'ExtraTreesClassifier': etc,
    'AdaBoostClassifier': ada,
    'BaggingClassifier': bc,
    'XGBClassifier': xgb
}
```

```python
models_name,ascore,roc,cross_val=[],[],[],[]
for name in models.keys():
    models[name].fit(x_train,y_train)
    models_name.append(name)
    y_pred=models[name].predict(x_test)
    ascore.append(round(accuracy_score(y_test,y_pred)*100,2))
    z=(cross_val_score(models[name],x,y,cv=10,scoring='accuracy').mean())
    cross_val.append(round(z*100,2))
    roc.append(round(roc_auc_score(y_test,y_pred)*100,2))
    print('################# ',name,' ## COMPLETED ##')
final=pd.DataFrame({'Model':models_name,'Accuracy Score':ascore,'ROC_AUC_Score':roc,'Cross Val Score':cross_val}).sort_values(by='Accuracy Score',ascending=False)
final
```

SVC is the best most model for this dataset with score of 87.41%
